```python
import pycorrelate as pyc
import numpy as np 
import matplotlib.pyplot as plt
import os
import scipy.signal as sig
%matplotlib inline
```

```python
# load data
data = np.load('./graham/flow_a-l_0.3-time_1500.0hr_CAT.npy')
n,m,t = data.T
```

```python
ms = sig.resample(m,int(m.size/0.1))
```

```python
s = 0.05
T = np.random.exponential(scale=s,size=int(50/s)).cumsum()
Y = np.sin(T)
```

```python
Y1 = sig.resample(Y,int(Y.size/0.1))
T1 = np.linspace(T.min(),T.max(),int(Y.size/0.1))
```

```python
plt.scatter(T,Y,color='blue')
plt.scatter(T1,Y1,color='red')
```

```python
auto = pyc.ucorrelate(Y1,Y1)
```

```python
plt.plot(auto)
```
